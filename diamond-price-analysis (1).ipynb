```python
# Import Necessary Libraries
import numpy as np # linear algebra
import pandas as pd # data processing
import matplotlib.pyplot as plt
import seaborn as sns

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
# Load the Dataset
df = pd.read_csv('/kaggle/input/diamond-dataset/diamonds.csv')
```

```python
# Display the First and Last Few Rows of the Dataset
df
```

```python
#Dataset Data Type 
df.info()
```

```python
#Description of the Data in the DataFrame
df.describe()
```

```python
# Create Scatter Plot to Show the Relationship Between Carat Weight and Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='carat', y='price', data=df, alpha=0.5)
plt.title('Diamond Price vs. Carat Weight')
plt.xlabel('Carat Weight')
plt.ylabel('Price ($)')
plt.show()
```

```python
# Create Box Plot to Visualize Price Distribution Across Different Colors
plt.figure(figsize=(10, 6))
sns.boxplot(x='color', y='price', data=df, order=['D', 'E', 'F', 'G', 'H', 'I', 'J'])
plt.title('Diamond Price Distribution by Color')
plt.xlabel('Color')
plt.ylabel('Price ($)')
plt.show()
```

```python
# Create Box Plot to Visualize Price Distribution Across Different Clarity Grades
plt.figure(figsize=(10, 6))
sns.boxplot(x='clarity', y='price', data=df, order=['IF', 'VVS1', 'VVS2', 'VS1', 'VS2', 'SI1', 'SI2', 'I1'])
plt.title('Diamond Price Distribution by Clarity')
plt.xlabel('Clarity')
plt.ylabel('Price ($)')
plt.show()
```

```python
# Calculate Correlation Matrix
corr_matrix = df[['carat','depth','table','price','x','y','z']].corr()
corr_matrix
```

```python
# Plot Correlation Heatmap
plt.figure(figsize=(10, 10))
sns.heatmap(corr_matrix, annot=True, cmap='mako',linewidth=.5, fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
